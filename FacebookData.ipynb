```python
import matplotlib.pyplot as plt
import networkx as nx
import requests
import random
import re
import os
import gzip

#get download
def download(url, to):
    response = requests.get(url)
    directory = ''.join(re.split('[/\\\]', to)[:-1])
    os.makedirs(directory, exist_ok=True)
    with open(to, "wb") as f:
        f.write(response.content)
        return to


def gz_unzip(file, out=None):
    if out is None:
        out = file.rstrip('.gz')

    with gzip.open(file, 'rb') as f_in:
        with open(out, 'wb') as f_out:
            f_out.write(f_in.read())
            return out

directory = r'C:\Users\Kristoffer\Downloads\facebook'

zip_file = download('https://snap.stanford.edu/data/facebook_combined.txt.gz', to='temp/facebook.txt.gz')
print(zip_file)
txt_file = gz_unzip(zip_file)
print(txt_file)

list_dir =  r'C:\Users\Kristoffer\Desktop\facebook_combined.txt'
print(directory)
#file_list = [open("{}\{}".format(directory, filename)) for filename in os.listdir(directory)]
print(f'version is {nx.__version__}')

#print(type(file_list))
#print(len(file_list))

fh = open(list_dir, 'rb')
G = nx.read_edgelist(fh)
fh.close()




#for value in file_list:
#    print(value)
#    G.add_node(value)


#antal noder
print(G.number_of_nodes())

#antal edges
print(G.number_of_edges())

#average degree
#ikke sikker på at denne metode virker men det kan også gøre manuelt
print('here is the dict ', dict(G.degree()))
print(nx.average_degree_connectivity(G))
degrees = dict(G.degree())
print(f'values are {degrees.values()}')
sum_of_edges = sum(degrees.values())
average_degree = (sum_of_edges / G.number_of_nodes())
print(average_degree)



#shorthand for biggest node
print(G)
biggest_node = max(dict(G.degree()).items(), key = lambda x : x[1])

print(f'biggest node is {biggest_node}')



#finding largest nodes
biggest_list = sorted(dict(G.degree()).items(), key = lambda x : x[1], reverse=True)
print(biggest_list[:5])

str = 'Alphabetical'


def scrabble(str_):
   return ''.join(random.sample(str_, len(str_)))


scrabble_str = scrabble(str)
print(scrabble_str)


nx.draw(G)
plt.show()





r'C:\Users\Kristoffer\PycharmProjects\numpy_practise'
'https://snap.stanford.edu/data/facebook.tar.gz'
```
